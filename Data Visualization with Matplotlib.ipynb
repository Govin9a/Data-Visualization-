```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetic_data = pd.read_csv('diabetic_data.csv')
Mapping_data = pd.read_csv('IDs_mapping.csv')
breast_cancer_data = pd.read_csv( 'breast_cancer.csv')
arrhythmia_data = pd.read_csv('arrhythmia.csv')
```

```python
diabetic_data
#using diabetic data to create the visulization
```

```python
# line graph using seaborn
sns.lineplot(data=diabetic_data, x= 'admission_type_id', y = 'time_in_hospital').set(title = 'Admission type VS Time in Hospital')
plt.show()
```

The time in hospital changes based on the type of admission. 

```python
#Ploting Barchart using seaborn
sns.barplot(data=diabetic_data, x= 'num_procedures', y = 'num_medications').set(title = 'Number of Procedures VS Number of Medications') #number of procedures  vs number of medication
plt.show()
```

The number of medication increases as the number of procedure increases. 

```python
#Boxplot visulization using seaborn
sns.boxplot(data=diabetic_data, x= 'gender', y = 'time_in_hospital').set(title = 'Gender VS Time in Hospital') #gender vs time in hospital
plt.show()
```

The mean hospital time span for female (4) is higher than male (3)

```python
#Bivariate plot using seaborn
sns.histplot(data=diabetic_data, x= 'num_medications', y = 'time_in_hospital').set(title = 'Number of medication vs Time in Hospital') #Number of medication vs time in hospital
sns.kdeplot(data=diabetic_data, x= 'num_medications', y = 'time_in_hospital')
plt.show()
```

The number of medication increases as the time spent in hospital increases.
